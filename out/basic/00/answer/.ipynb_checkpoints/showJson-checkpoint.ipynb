```python
import pandas as pd
test_pd = pd.DataFrame(data={'id':list(test_data.keys()), 'phrase_pred':list(test_data.values())}, columns=['id', 'phrase_pred'])
test_pd.head(10)
```

```python
col_names = ['id', 'Query', 'Url', 'Answer', 'AnswerTokenList', 'ParaseSpan', 'phrase', 'Label']
origin_dev_data = pd.read_csv('/home/t-honli/data/squad/EQnA_Highlighting_Test.tsv', header=None, sep='\t', names=col_names, dtype=str)
origin_dev_data.head(5)
```

```python
test_origin = origin_dev_data[origin_dev_data['Label'] == '1']
```

```python
test_pd_merge = pd.merge(test_origin, test_pd, on='id', how='left')
test_pd_merge.head(5)
```

```python
test_pd_merge_label = test_pd_merge.pop('Label')
```

```python
test_pd_merge.insert(len(test_pd_merge.columns), 'Label', test_pd_merge_label)
```

```python
# 1. How many rows that have the same query, url, answer, label?( 1 context -> 1 quesiotn -> multi phrase)
import hashlib
def GetHashCode(row):
    hash = hashlib.md5()
    hash.update((row['Query'] + ' ' + row['Url'] + ' ' + row['Answer']).encode('utf-8'))
    return hash.hexdigest()
hash_code = test_pd_merge.apply(GetHashCode, axis=1 )
hash_code.head(5)
```

```python
test_pd_merge.insert(len(test_pd_merge.columns), 'HashCode', hash_code)
```

```python
test_pd_merge.head(10)
```

```python
from collections import Counter
count = Counter(test_pd_merge['HashCode'].tolist())
dultiple_cnt = 0
dultiple_sum = 0
for idx, cnt in list(count.items()):
    if cnt > 1:
        dultiple_cnt += 1
        dultiple_sum += cnt
print ('dultiple num', dultiple_cnt)
print ('dultiple sum', dultiple_sum)
print ('extra num', dultiple_sum - dultiple_cnt)
print ('totoal num', len(test_pd_merge))
```
